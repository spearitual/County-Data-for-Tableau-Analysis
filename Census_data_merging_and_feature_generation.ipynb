## Fill in county and state names based on FIPS wherever they are missing

```python
df_bfill = df.groupby('FIPS').bfill()[['date','county','state','state_abbr','county_state']]
df_bfill
```

```python
df.county = df_bfill.county
df.state = df_bfill.state
df.state_abbr = df_bfill.state_abbr
df.county_state = df_bfill.county_state
df
```

```python
df_ffill = df.groupby('FIPS').ffill()[['date','county','state','state_abbr','county_state']]
df_ffill
```

```python
df.county = df_ffill.county
df.state = df_ffill.state
df.state_abbr = df_ffill.state_abbr
df.county_state = df_ffill.county_state
df
```

## Remove unneeded rows

```python
#Remove Peurto Rico
df = df.loc[(df.FIPS < '72000')]
df
```

```python
#Remove two placeholder counties
df = df.drop(list(df.loc[(df.land_area_sqmi.isna())].index))
```

## Feature generation

### Housing rate

```python
df['housing_rate'] = round(df.population_total/df.housing_units,2)
df.loc[(df.date == '2001-07-01')]
```

### Population density

```python
df['pop_density'] = round(df.population_total/df.land_area_sqmi,3)
df.loc[(df.date == '2001-07-01')]
```

### Annual population change

```python
df['pop_shift'] = df.groupby('FIPS').shift(-1)['population_total']
df.loc[(df.date == '2001-07-01')]
```

```python
df['pop_change'] = round(((df.pop_shift-df.population_total)/df.population_total)*100,2)
df.loc[(df.date == '2001-07-01')]
```

```python
df = df.drop('pop_shift',axis=1)
df.loc[(df.date == '2001-07-01')]
```

### Annual income change

```python
df = df.rename(columns={'average_income':'income','uninflated_income':'real_income'})
```

```python
df['income_shift'] = df.groupby('FIPS').shift(-1)['income']
df['income_change'] = round(((df.income_shift-df.income)/df.income)*100,2)
df = df.drop('income_shift',axis=1)
df
```

```python
df['real_income_shift'] = df.groupby('FIPS').shift(-1)['real_income']
df['real_income_change'] = round(((df.real_income_shift-df.real_income)/df.real_income)*100,2)
df = df.drop('real_income_shift',axis=1)
df
```

### Quantiles

```python
quantiles = pd.qcut(df.population_total, 20, labels=False)
df = df.assign(pop_quantiles=quantiles.values)
```

```python
quantiles = pd.qcut(df.pop_density, 20, labels=False)
df = df.assign(pop_density_quantiles=quantiles.values)
```

```python
quantiles = pd.qcut(df.real_income, 20, labels=False)
df = df.assign(real_income_quantiles=quantiles.values)
```

```python
df.pop_quantiles = df.pop_quantiles + 1
df.pop_density_quantiles = df.pop_density_quantiles + 1
df.real_income_quantiles = df.real_income_quantiles + 1
```

```python
df.loc[(df.pop_density_quantiles == 20)]
```

### Housing rate divergence

```python
dfreg = df.loc[~(df.housing_rate.isna())]
dfreg = dfreg.loc[~(df.median_age_total.isna())]
dfreg
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(dfreg[['median_age_total']],dfreg.housing_rate)
r2_score = lr.score(dfreg[['median_age_total']],dfreg.housing_rate)
r2_score
```

```python
dfreg['housing_rate_divergence']= dfreg.housing_rate - lr.predict(dfreg[['median_age_total']])
dfreg
```

```python
dfreg.housing_rate_divergence
```

```python
df = df.join(dfreg.housing_rate_divergence, how='outer')
df
```
