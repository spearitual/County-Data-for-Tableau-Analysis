1. Convert YEAR to datetime

### Covert YEAR to datetime

```python
for old_FIPS, new_FIPS in FIPS_swapper.items():
    df['FIPS'] = df.FIPS.replace(old_FIPS, new_FIPS)
for old_name, new_name in name_swapper.items():
    df['CTYNAME'] = df.CTYNAME.replace(old_name, new_name)
```

```python
#Combine data for combined counties. Mean of medians won't be precisely accurate.
df = df.groupby(['FIPS','STNAME','CTYNAME','YEAR']).agg({'POPESTIMATE':'sum','POPEST_MALE':'sum','POPEST_FEM':'sum','POPESTIMATE':'sum','MEDIAN_AGE_TOT':'mean','MEDIAN_AGE_MALE':'mean','MEDIAN_AGE_FEM':'mean'}).reset_index()
```

### Create Percent Male column

```python
df['percent_male']=df['POPEST_MALE']/(df['POPEST_MALE']+df['POPEST_FEM'])
df.head()
```
